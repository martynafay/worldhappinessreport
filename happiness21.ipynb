### explaination of the columns
**country**: the country in question

**region**: the region that the country belongs to

**gdp_per_capita**: economy; individual rank their quality of life based on the amount they earn

**social_support**: "if you're in trouble, is there someone you can call on?" 
                aka quality of friend and social life

**health_life_expectancy**: ranking healthcare availability and average life expectancy in the country

**freedom**: how much an individual is able to conduct themself based on their free will

**generosity**: how much their country is involved in peacekeeping and global aid

**corruption_level**: how corrup the government is

**dystopia_residual**: dystopia happiness score (1.85) 
_(aka the rank of a hypothetical country that has the lowest happiness score ever; yk like a dystopia; hunger games)_ plus the residual value
of each country _(unexplainable happiness/sadness; if you have a high residual you're unexplainably happy, if you have a low residual you're unexplainable sad)_

```python
#prediction target
y = happiness21.happiness_score
```

```python
#features to use for prediction
happiness21_features = ['region', 'gdp_per_capita', 'social_support', 'healthy_life_expectancy', 'freedom',
 'generosity', 'corruption_level', 'dystopia_residual']
X = happiness21[happiness21_features]
```

```python
#getting a look at my predictions and features
y.head()
```

```python
X.head()
```

```python
#to create training and test splits
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
#before you can train your data, you need to scale it so everything is on the same measurement
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
#transforming is centering and scaling
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#using a regressor to train my data
#a regressor is used because we are dealing with continous values
from sklearn.neural_network import MLPRegressor
regr = MLPRegressor(hidden_layer_sizes = (100, 100, 100,), random_state=1, max_iter=10000) #the more hidden layers and iterations, the more acccurate it becomes
regr.fit(X_train, y_train.values.ravel())
```

```python
predictions = regr.predict(X_test)
```

```python
#to visually compare the test happiness score to what was predicted
print(y_test)
print(predictions)
```

```python
regr.score(X_test, y_test) #this changes everytime, but for the most part it's 80% accurate
```

```python
regr.score(X_train, y_train) #the training set being 100% doesn't surprise me
```
